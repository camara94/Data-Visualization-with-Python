```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_can = pd.read_csv('canada.csv')
df_can.head(2)
```

```python
df_can['Unnamed: 0'];
```

```python
years = list(map(str, range(1980, 2014)))
df_can.index = df_can['Unnamed: 0']
df_can.loc[['Guinea'], years].transpose().plot(kind='box', figsize=(15, 20))
plt.title('Box plot of Guinea immigrants from 1980 to 2013')
plt.ylabel('Number of immigrants')
```
