```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_can = pd.read_csv('canada.csv')
df_can.head(2)
```

```python
df_can.rename(columns={"Unnamed: 0": "Country"}, inplace=True)
```

```python
df_can.index = df_can.iloc[:, 0]
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)
```

```python
top_5_countries = df_can.head()
```

```python
top_5_countries = top_5_countries[years].transpose()
top_5_countries
```

```python
top_5_countries.plot(kind='line')
plt.title('Immigration trend of top 5 countries')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
```
