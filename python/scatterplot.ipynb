```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_can = pd.read_csv('canada.csv')
df_can.head(3)
```

```python
years = list(map(str, range(1980, 2014)))
df_scatter = df_can.loc[:, years].transpose()
new_data = pd.DataFrame()
new_data['years'] = df_scatter.index  
new_data['Total'] = df_can['Total'][:]
```

```python
new_data.head()
```

```python
new_data.plot(kind='scater', x='years', y='Total')
```
